```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
from sklearn.preprocessing import OneHotEncoder
```

```python
# Loading the dataset
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True,axis=1)
data
```

```python
# Encode Gender
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
# One hot Encode Geography
oh = OneHotEncoder(sparse=False)
geo_encoder = oh.fit_transform(data[['Geography']])
geo_encoder
```

```python
type(geo_encoder)
```

```python
geo_enoded_df = pd.DataFrame(geo_encoder,columns=oh.get_feature_names_out(['Geography']))
```

```python
geo_enoded_df
```

```python
# Combining the columns 
data = pd.concat([data.drop('Geography',axis=1),geo_enoded_df],axis=1) # axis 1 = column
data
```

```python
# save encoder,scaler
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('oh.pkl','wb') as file:
    pickle.dump(oh,file)
```

```python
data.head()
```

```python
# Divide the dataset
x = data.drop('Exited',axis=1)
y = data['Exited']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Scaling the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Pickling scaler
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
x_train.shape[1]
```

```python
# Building ANN model
model = Sequential(
    [
        Dense(64,activation='relu',input_shape=(x_train.shape[1],)), # First hidden layer connect with input layer
        Dense(32,activation='relu'),
        Dense(1,activation='sigmoid') # Output layer
    ]
)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Setup tensorboard
log_dir = 'logs/fit/' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python

tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
# Setup EarlyStopping
early_stopping_callback = EarlyStopping(monitor='val_loss',patience = 5, restore_best_weights=True)
```

```python
# Training the model
history = model.fit(
    x_train,y_train,
    validation_data = (x_test,y_test),
    epochs = 100,
    callbacks = [tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
